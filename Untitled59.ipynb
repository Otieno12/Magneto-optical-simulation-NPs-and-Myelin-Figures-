```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

fig = plt.figure(figsize=(10,4))

# -------------------------
# LEFT: Planar waveguide
# -------------------------
ax1 = fig.add_subplot(121, projection='3d')

# Dimensions
L, W, H = 3, 2, 1
core_thickness = 0.3

# Helper function to draw cuboid
def draw_box(ax, x, y, z, dx, dy, dz, color, alpha=0.9):
    vertices = np.array([
        [x,y,z],[x+dx,y,z],[x+dx,y+dy,z],[x,y+dy,z],
        [x,y,z+dz],[x+dx,y,z+dz],[x+dx,y+dy,z+dz],[x,y+dy,z+dz]
    ])
    faces = [
        [vertices[j] for j in [0,1,2,3]],
        [vertices[j] for j in [4,5,6,7]],
        [vertices[j] for j in [0,1,5,4]],
        [vertices[j] for j in [2,3,7,6]],
        [vertices[j] for j in [1,2,6,5]],
        [vertices[j] for j in [0,3,7,4]]
    ]
    ax.add_collection3d(Poly3DCollection(faces, facecolors=color, alpha=alpha))

# Bottom cladding n2
draw_box(ax1, 0, 0, 0, L, W, 0.35, 'cyan')
# Core n1
draw_box(ax1, 0, 0, 0.35, L, W, core_thickness, 'gold')
# Top cladding n3
draw_box(ax1, 0, 0, 0.35 + core_thickness, L, W, 0.35, 'turquoise')

ax1.text(1.4, 0.8, 0.5, r'$n_1$', fontsize=12)
ax1.text(1.4, 0.8, 0.15, r'$n_2$', fontsize=12)
ax1.text(1.4, 0.8, 0.85, r'$n_3$', fontsize=12)

ax1.set_title('Planar (slab) waveguide\n$(n_1 > n_2, n_3)$', fontsize=11)
ax1.set_axis_off()
ax1.set_box_aspect((L,W,H))

# -------------------------
# RIGHT: Cylindrical fiber
# -------------------------
ax2 = fig.add_subplot(122, projection='3d')

z = np.linspace(0, 3, 100)
theta = np.linspace(0, 2*np.pi, 50)
theta, z = np.meshgrid(theta, z)

# Core
r1 = 0.3
x1 = r1 * np.cos(theta)
y1 = r1 * np.sin(theta)

# Cladding
r2 = 0.8
x2 = r2 * np.cos(theta)
y2 = r2 * np.sin(theta)

ax2.plot_surface(x2, y2, z, color='turquoise', alpha=0.6)
ax2.plot_surface(x1, y1, z, color='gold', alpha=1.0)

ax2.text(0.1, 0.1, 1.5, r'$n_1$', fontsize=12)
ax2.text(0.7, 0.7, 1.5, r'$n_2$', fontsize=12)

ax2.set_title('Cylindrical waveguide\n$(n_1 > n_2)$', fontsize=11)
ax2.set_axis_off()
ax2.set_box_aspect((1,1,3))

plt.tight_layout()
plt.show()
```
