```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
fig, ax = plt.subplots(figsize=(8, 5))

# Laser beam shape
z = np.linspace(-5, 5, 400)
beam_width = 0.2 + 0.05 * z**2

ax.fill_between(z, beam_width, -beam_width,
                color='lightblue', alpha=0.6, label='Laser beam')

# Beam waist
ax.plot([0, 0], [-0.25, 0.25], color='blue', linewidth=3)
ax.text(0.05, 0.3, 'Beam waist', fontsize=10)

# Myelin tube (cylindrical object)
circle = plt.Circle((0, 0), 0.35, color='orange', alpha=0.9)
ax.add_patch(circle)
ax.text(-0.35, -0.65, 'Myelin tube', fontsize=10)

# Gradient force arrow
ax.arrow(-1.5, 0, 1.1, 0,
         head_width=0.08, head_length=0.15,
         fc='red', ec='red')
ax.text(-1.5, 0.1, 'Gradient force', color='red', fontsize=10)

# Scattering force arrow
ax.arrow(0.5, 0, 1.2, 0,
         head_width=0.08, head_length=0.15,
         fc='green', ec='green')
ax.text(0.7, 0.1, 'Scattering force', color='green', fontsize=10)

# Labels
ax.text(-4.5, 1.2, 'Focused laser beam', fontsize=11)

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-5, 5)
ax.set_ylim(-2, 2)
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch

def draw_panel(ax, offset_force=True):
    # Coordinate system
    z = np.linspace(-4, 4, 400)
    beam = 0.4 + 0.05 * z**2

    # Laser intensity profile
    ax.fill_between(z, beam, -beam, color='red', alpha=0.25)
    ax.plot(z, beam, color='darkred', linewidth=1)
    ax.plot(z, -beam, color='darkred', linewidth=1)

    # Beam axis
    ax.plot([0, 0], [-3, 3], linestyle='--', color='black', linewidth=1)

    # Trapped particle
    x_center = -0.5 if offset_force else 0
    particle = Circle((x_center, 0), 0.45, color='steelblue', zorder=5)
    ax.add_patch(particle)

    # Gradient force arrows
    ax.arrow(x_center - 1.0, 0.8, 0.8, -0.5,
             head_width=0.12, head_length=0.18,
             fc='black', ec='black')
    ax.text(x_center - 1.1, 1.0, r'$F_1$', fontsize=10)

    ax.arrow(x_center + 1.0, 0.8, -0.8, -0.5,
             head_width=0.12, head_length=0.18,
             fc='black', ec='black')
    ax.text(x_center + 0.9, 1.0, r'$F_2$', fontsize=10)

    # Net force
    ax.arrow(x_center, 0.2, 0.0, -1.2,
             head_width=0.15, head_length=0.25,
             fc='black', ec='black', linewidth=2)
    ax.text(x_center + 0.05, -1.2, r'$F_{\mathrm{net}}$', fontsize=10)

    # Labels
    ax.text(-3.8, -2.6, '1', fontsize=10)
    ax.text(3.4, -2.6, '2', fontsize=10)
    ax.text(-2.2, -3.4, 'intensity profile', fontsize=10)
    ax.text(1.2, -3.4, 'laser light in', fontsize=10)

    # Formatting
    ax.set_xlim(-4, 4)
    ax.set_ylim(-3.5, 3.5)
    ax.set_aspect('equal')
    ax.axis('off')


# Create figure
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Panel (a): Off-axis particle (unstable / restoring forces)
draw_panel(axes[0], offset_force=True)
axes[0].set_title('(a)', loc='left')

# Panel (b): On-axis particle (stable trapping)
draw_panel(axes[1], offset_force=False)
axes[1].set_title('(b)', loc='left')

plt.tight_layout()
plt.show()
```
